# Matplotlib Pt. 1

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

y is x^2

## Functional Method

```python
plt.plot(x, y)
# plt.show() is done within jupyter notebooks by printing the outcell and printing the graph separately
```

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
# plt.subplot(number of rows, number of columns, plot number you're referring to)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## Object-Oriented Method

```python
fig = plt.figure()
# fig.add_axes([% from left of the axes, bottom of axes, width of axes, height of axes])
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
```

```python
fig = plt.figure() # made an empty figure
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # pass a list of axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Larger Plot')
axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

## Matplotlib Pt. 2

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2) # subplots can specify the number of rows and columns
# plt.tight_layout() fixes overlaps 
for current_ax in axes:
    current_ax.plot(x, y)
```

```python
axes # is just an array (iterable)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# we can also index-call the axes

axes[0].plot(x, y)
axes[0].set_title('First Plot')
axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

## Figure Size and DPI

```python
fig = plt.figure(figsize = (8, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig
```

```python
fig.savefig('my_picture.png') # can specify dpi here
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label = 'X Squared')
ax.plot(x, x**3, label = 'X Cubed')

ax.legend(loc = 10) # use documentation for numbers to use to move the legend. It can take tuples as well.
```

# Matplotlib Pt. 3

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = '#FF8C00', linewidth = 3, alpha = 0.5, linestyle = '--')
# can customize line color, line thickness, transparency, line style
# linewidth can be replaced with lw, linestyle can be replaced with ls, and they would still work
```

```python
len(x) #length of array of x
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'purple', linewidth = 3, alpha = 0.5, linestyle = '-', marker = 'o', markersize = 10, markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green')
# markers will denote the points of x vs y
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'purple', lw = 2, ls = '--')

ax.set_xlim([0, 1]) # sets lower and upper bound of x axis
ax.set_ylim([0, 2]) # sets lower and upper bound of y axis
```
